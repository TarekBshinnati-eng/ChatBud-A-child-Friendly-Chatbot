# Gemma 3 4B-IT: Download to Drive and Load from Local Folder

This notebook downloads the `google/gemma-3-4b-it` model once into Google Drive. Future sessions can load the model directly from Drive instead of redownloading.
This notebook is stored at the repository root (main branch) as `gemma3_drive_setup.ipynb` for easy access without any subfolders.


```python
import os
from google.colab import drive

drive.mount("/content/drive")
```

```python
BASE_DRIVE_DIR = "/content/drive/MyDrive/models"
MODEL_DIR = os.path.join(BASE_DRIVE_DIR, "gemma-3-4b-it")

os.makedirs(MODEL_DIR, exist_ok=True)
```

```python
!pip install -q huggingface_hub
from huggingface_hub import snapshot_download

HF_TOKEN = os.environ.get("HF_TOKEN", None)
# Optionally set HF_TOKEN manually if needed:

repo_id = "google/gemma-3-4b-it"

snapshot_download(
    repo_id=repo_id,
    local_dir=MODEL_DIR,
    local_dir_use_symlinks=False,
    token=HF_TOKEN  # can be None if no token required or env is configured
)

print("Downloaded model snapshot to:", MODEL_DIR)
```
