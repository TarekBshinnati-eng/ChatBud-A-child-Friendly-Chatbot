# Export Child-QA test split to CSV

This notebook loads the `chaitanyareddy0702/Child-QA-dataset` from Hugging Face, takes the **test** split (questions and answers), and saves it to `prompts.csv`.

```python
!pip install -q datasets pandas
```

```python
from datasets import load_dataset
import pandas as pd
```

```python
dataset_id = "chaitanyareddy0702/Child-QA-dataset"

# load only the test split (should contain 244 rows)
test_ds = load_dataset(dataset_id, split="test")
print(test_ds)  # quick sanity check of columns and size
```

```python
df = test_ds.to_pandas()

# ensure a clean integer id column starting at 1
df.insert(0, "id", range(1, len(df) + 1))

# keep only id, Question, Answer in this order if extra columns exist
keep_cols = [col for col in ["id", "Question", "Answer"] if col in df.columns]
df = df[keep_cols]

df.head()
```

```python
output_path = "prompts.csv"
df.to_csv(output_path, index=False, encoding="utf-8")
print(f"Saved {len(df)} rows to {output_path}")
```

## Result

The file `prompts.csv` has been created in the current directory and contains all rows from the **test** split of `chaitanyareddy0702/Child-QA-dataset` with their questions and answers.
